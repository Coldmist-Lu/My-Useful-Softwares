```python
import pandas as pd
import random
import os
```

```python
# 初始化字典列表
def Initiating_wordlist():

    INITIATING_WORD_NUMBER = 10 # 初始化单词数量

    print("现在开始初始化")
    wordlist_name = input("请输入词库名：")              

    eng = []
    chi = []
    print("现在开始录入10个单词：")

    for i in range(INITIATING_WORD_NUMBER):
        print("第{}个单词：".format(i))
        eng.append(input("请输入英文："))
        chi.append(input("请输入中文："))

    wordlist = pd.DataFrame(pd.DataFrame(list(zip(eng, chi)), columns = ['English', 'Chinese']))

    wordlist.to_csv(wordlist_name + '.csv', encoding="gbk")
    print("词库已被保存在：{}".format(os.path.abspath("")))

    print("词库数据初始化完成！")
```

```python
class ManipulateWordList():
    """
    单词数据和操作
    """
    
    def __init__(self, file_name="Dictionary", file_folder=""):
        self.ListName = file_name
        self.file_path = file_folder + file_name + '.csv'
        self.WordList = pd.read_csv(self.file_path, encoding="gbk") # 加载单词数据
        self.WordList = self.WordList.iloc[:, 1:]
        
    def find_words(self, eng):
        return self.WordList[self.WordList['English'].isin([eng])].index.tolist() # 找单词
        
    def add_words(self):
        while True:
            print("下面输入一个新单词：")
            eng = input("请输入英文：")
            chi = input("请输入中文：")
            if eng == '-' or chi == "-":
                break
            else:
                self.WordList = self.WordList.append({'English':eng, 'Chinese':chi}, ignore_index=True)
                
        self.WordList.to_csv(self.file_path, encoding="gbk")
        print("添加单词成功！")
        return self.WordList
        
    def delete_words(self):
        while True:
            print("下面删除一个单词：")
            eng = input("请输入英文：")
            if eng == '-':
                break
            else:
                idx = self.find_words(eng) # 找单词
                if idx:
                    self.WordList = self.WordList.drop(idx) # 删除
                else:
                    print("无该单词！")
                
        self.WordList.to_csv(self.file_path, encoding="gbk")
        print("删除单词成功！")
        return self.WordList
        
    def get_list(self, start=0, end=None):
        if end == None:
            end = len(self.WordList)
        return self.WordList.iloc[start:end]
    
    def recite_words(self, num = 20, end = None):
        print("现在开始背单词：")
        word_ls = self.WordList['English'].tolist()
        
        if end:
            word_ls = word_ls[:end]
        
        if len(word_ls) < num:
            print("单词数少于{0}个，自动减少到{1}个".format(num, len(word_ls)))
            num = len(word_ls)
        
        rc_words = random.sample(word_ls, num)
        random.shuffle(rc_words)
        
        score = 0
        for word in rc_words:
            print("请回忆：{}".format(word))
            res = eval(input("请输入：(0:忘记;1:模糊;2:掌握)"))
            idx = self.find_words(word)[0]
            print("中文意思是：{}".format(self.WordList.iloc[idx, 1]))
            score += res
            
        print("正确率：{:.1f}%".format(score / (2*num) * 100))
        
    def menu(self):
        print("欢迎来到 Sirius 的背单词软件，当前词库为{}，请选择你需要使用的内容:".format(self.ListName))
        print("0:记忆单词; 1:添加单词; 2:删除单词; 3:背诵单词 4:退出")
        menu_num = eval(input())
        
        if menu_num == 0:
            idx = input("请输入单词索引：")
            ls = idx.split(" ")
            if len(ls) == 0:
                return self.get_list()
            elif len(ls) == 1:
                return self.get_list(start=eval(ls[0]))
            else:
                return self.get_list(start=eval(ls[0]), end=eval(ls[1]))
            
        elif menu_num == 1:
            return self.add_words()
            
        elif menu_num == 2:
            return self.delete_words()
        
        elif menu_num == 3:
            num = eval(input("请输入记忆单词数量:"))
            end_loc = eval(input("请输入考查结尾位置(若全部考查输入-1):"))
            if end_loc == -1:
                self.recite_words(num)
            else:
                self.recite_words(num, end_loc)
            
        else:
            return
```

```python
mwl = ManipulateWordList("IELTS_Dictionary")
```
